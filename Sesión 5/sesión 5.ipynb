```python
cuentas=pd.read_csv("data/accounts.csv")
```

```python
cuentas
```

```python
sns.scatterplot(data=cuentas, x="date_opened",y="total")
```

```python
account_offices=pd.read_csv("data/account_offices.csv")
account_offices
```

```python
cuenta_global=pd.merge(cuentas,account_offices, on="id")
cuenta_global
```

```python
cuenta_global["office"]=="Tokyo"
```

```python
filtro_tokyo=cuenta_global[cuenta_global["office"]=="Tokyo"]
filtro_tokyo
```

```python
cuenta_global.loc[filtro_tokyo, 'total']/104
```

```python
cuenta_global.loc[filtro_tokyo, "total"]=cuenta_global.loc[filtro_tokyo,"total"]/104
```

```python
sns.scatterplot(data=cuenta_global, x="date_opened", y="total")
```

```python
cuenta_global['total_usd'] = np.where(cuenta_global['office'] == 'Tokyo',cuenta_global['total'] / 104,cuenta_global['total'])
print(cuenta_global)
```

```python
sns.scatterplot(data=cuenta_global, x="date_opened", y="total_usd")
```
