# Extra: Concatenando DataFrames

```python
data=[[1,2,3],[4,5,6],[7,8,9]] #Creando DF por medio de arrays
data
```

```python
df1=pd.DataFrame(data=data)
df1
```

```python
index=['Linea' +str(i) for i in range(3)]
index
```

```python
df1=pd.DataFrame(data=data, index=index)
df1
```

```python
columns=['Columna' +str(i) for i in range(3)]
columns
```

```python
df1=pd.DataFrame(data=data, index=index, columns=columns)
df1
```

```python
data={'Columna0':{'Linea0':1,'Linea1':4,'Linea2':7},
      'Columna1':{'Linea0':2,'Linea1':5,'Linea2':8},
      'Columna2':{'Linea0':3,'Linea1':6,'Linea2':9}}
data   #Creando DF por medio de diccionarios
```

```python
df2=pd.DataFrame(data)
df2
```

```python
data2=[(1,2,3),(4,5,6),(7,8,9)]
data2 #Creando DF por medio de tuplas
```

```python
df3=pd.DataFrame(data2, index=index,columns=columns)
df3 
```

```python
df1[df1>0]='A'  #Aplicando filtros y cambio de valores: df1
df1
```

```python
df2[df2>0]='B'  #Aplicando filtros y cambio de valores: df2
df2
```

```python
df3[df3>0]='C'  #Aplicando filtros y cambio de valores: df3
df3
```

```python
df4=pd.concat([df1,df2,df3]) #Concatenación hacia abajo
df4
```

```python
df5=pd.concat([df1,df2,df3],axis=1)  #Concatenación a la derecha
df5
```
