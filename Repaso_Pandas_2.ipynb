# 7. Selecciones y Frecuencias

```python
db=pd.read_csv('alquiler_residencial.csv',sep=';')
db.head(10)
```

```python
# Seleccionar los inmuebles clasificados con tipo: Departamento
seleccion=db[db['Tipo']=='Departamento']
seleccion
```

```python
#Frecuencia de criterio anterior
n1=seleccion.shape[0]
n1
```

```python
#Seleccionar inmueles clasificados con tipo: Casa, Casa en condominio y Casa de villa
seleccion2= db[(db['Tipo'] =='Casa') | (db['Tipo'] =='Casa en condominio') |(db['Tipo'] =='Casa de villa')]
seleccion2
```

```python
#Frecuencia de criterio anterior
n2=seleccion2.shape[0]
n2
```

```python
#Seleccionar los inmuebles con área entre 60 y 100 metros cuadrados incluyendo limites
seleccion3=db[(db['Area']>=60 ) & (db['Area']<=100)]
seleccion3
```

```python
#Frecuencia de anterior criterio
n3=seleccion3.shape[0]
n3
```

```python
#Seleccionar los inmuebles que tengan por lo menos 4 cuartos y alquiler menor de $2000.00
seleccion4=db[(db['Cuartos']>=4) & (db['Valor']<2000 )]
seleccion4
```

```python
#Frecuencia de anterior criterio
n4=seleccion4.shape[0]
n4
```

```python
#Respuestas de frecuencias
print("Nº de inmuebles clasificados con tipo 'Departamento': ", n1)
print("Nº de inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa': ", n2)
print("Nº de inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los limites: ", n3)
print("Nº de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00: ", n4)
```
